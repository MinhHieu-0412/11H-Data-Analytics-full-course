## format  
`format` - dùng để **chèn giá trị của một biến vào chuối ký tự**  
- `{}` là **vị trí cần chèn**, `format()` là **các biến cần chèn** vào vị trí `{}`  
- Như code dưới, ta thấy giá trị của `role` là **"Data Analyst"** sẽ được **chèn** vào vị trí `Role: {}` và giá trị của biến `skill` là **"Python"** sẽ được **chèn** vào vị trí `Skill` `Required: {}`

```python
#Thiết lập 2 biến str
role = "Data Analyst"
skill = "Python"

#Sử dụng format để chèn biến vào chuỗi ký tự
print("Role: {}; Skill Required: {}".format(role, skill))
print(f"Role: {role}; Skill Required: {skill}")
print("Role: %s; Skill Required: %s" % (role, skill))
```

## **for** 
- `for` - được dùng để **lặp** qua các phần tử của 1 tập hợp  
-  Như ví dụ dưới đây, `number` là **biến tạm thời** để nhận các giá trị trong `numbers` khi `for` hoạt động, ứng với mỗi giá trị khi lặp sẽ được **in ra ngoài** bằng `print(number)` và `print` **phải thụt lề** so với `for`

```python
#Thiết lập một biến numbers là list của 1 tập hợp số
numbers = [0, 1, 2, 3, 4, 5]
#number sẽ nhận lần lượt các giá trị trong list thông qua lệnh lặp for
for number in numbers:
    #Với mỗi giá trị nhận được sẽ in ra màn hình
    print(number)
```

- Ngoài ra `for` còn có thể áp dụng cho `string`
- Như ví dụ dưới đây, `characters` là một **chuỗi ký tự (string)** và lệnh `for` sẽ **lặp từng ký tự** và trả về cho `character`, mỗi ký tự `character` nhận sẽ được in ra màn hình

```python
#Thiết lập một biến characters là 1 kiểu dữ liếu string Python
characters = "Python"

#character sẽ nhận lần lượt các kí tự có trong chuỗi thông qua lệnh lặp for
for character in characters:
    
    #Với mỗi ký tự mà character nhận sẽ được in ra màn hình
    print(character)
```

- `for` còn có thể tạo 1 dãy số, ký tự trong 1 `list` như các ví dụ sau:

## Ứng dụng for qua ví dụ dưới đây
- Khởi tạo một `list` rỗng `qualifiedRoles`
- Khởi tạo một `list` gồm `dict` bên trong, có các **keys và values** tương ứng
- Khởi tạo một biến `mySkills` là một `list` gồm các giá trị `string`
- Dùng `for` lặp các giá trị trong `list` `jobRoles` qua biến `job` và tạo một biết `requiredSkills` là `key "skill"` trong các phần tử `dict` của `job`
- Tạo một biến `qualified` với giá trị là `True` `(boolean)` 
- Dùng for để **lặp các giá trị** trong biến `requiredSkills` qua biến `skill`, dùng `if` tạo điều kiện **nếu biến** `skill` **không có trong** `mySkills` thì `qualified` sẽ mang giá trị `False` và `break` lệnh `for`
- Đặt tiếp điều kiện `if` nếu `qualified` **không có** `False` **thì sẽ thêm giá trị của** `job` vào `list` `qualifiedRoles`

```python
#Tạo một list rỗng
qualifiedRoles = []

#Tạo một list gồm dict bên trong và vị trí công việc cũng như yêu cầu skill cảu công việc
jobRoles = [
    {
        "role" : "Data Analyst",
        "skill" : ["Excel", "SQL", "Python"]
    },
    {
        "role" : "Business Intelligence Analyst",
        "skill" : ["PowerBI", "SQL"]
    }
]
#Biến mySKills sẽ gồm các skills mà người dùng có
mySkills = ["Python", "SQL", "Excel", "Java Script"]

#Lặp các giá trị trong jonRoles vào biến job
for job in jobRoles:

    #Trong biến job, các values từ key "skill" sẽ được nhận bởi biến requiredSkills tức những skill bắt buộc 
    requiredSkills = job["skill"]

    #Khởi tạo một biến qualified và mang giá trị True (boolean)
    qualified = True

    #lặp các giá trị của requiredSkills vào biến skill
    for skill in requiredSkills:

        #Khởi tạo điều kiện nếu skill vừa lặp từ for ở trên không có trong mySkills thì qualified sẽ mang giá trị False và break for ngay khi phát hiện
        if skill not in mySkills:
            qualified = False
            break

    #Nếu qualified không mang giá trị False thì sẽ thêm giá trị của biến job vào list rỗng qualifiRoles
    if qualified:
        qualifiedRoles.append(job)
print(qualifiedRoles)

```

# **while**

- `while` sẽ tạo thành **1 vòng lặp** để kiểm tra lại điều kiện
- Như ví dụ dưới, `while a < 5:` cứ mỗi khi **a nhỏ hơn 5** thì `a` sẽ được in ra màn hình và thực hiện lệnh **tăng biến đêm** `a += 1`, sau đó kết quả sẽ **lại một lần nữa quay về** `while` **để duyệt**, nếu `a` vẫn nhỏ hơn 5 thì sẽ **tiếp tục vòng lặp cũ đến khi a lớn hơn 5 thì dừng** thực hiện

```python
#Khởi tạo một biến a và cho giá trị của nó bằng 1
a = 1
#Dùng vòng lặp while, mỗi khi a nhỏ hơn 5 sẽ print ra màn hình và áp dụng a += 1 sau đó quay lại while để tiếp tục duyệt
while a < 5:
    print(a)
    a += 1
```

## while kết hợp với list, dict như ví dụ sau
- Khởi tạo 1 `tuple` gồm các `keys` và `values` tương ứng
- khởi tạo biến `a` để làm **điều kiện dừng vòng lặp**
- Khởi tạo biến `b` để **truy cập vào các phần tử** `list`
- Khởi tạo biến `yearsList` **để ép** `dict` **về** `list`
- Bắt đầu vòng lặp trong khi `a < 5`, biến `key` và `years` sẽ lần lượt nhận `keys` và `values` của `list` (`dict` `yearExperience` đã bị ép thành `list`) thông qua việc `b` đã **truy cập vào các phần tử** có trong `list`
- từ đó in ra màn hình `key` và `years` tương ứng, lúc này `years` nhận giá trị 3 từ `list`
- Thực hiện phép **tăng biến đếm** cho `b`, `b += 1` lúc này `b` nhận giá trị `1` (tức khi truy cập vào `list` sẽ là **phần tử thứ 2**) và bắt đầu **quay lại** `while` để tiếp tục vòng lặp cho đến khi `years` lớn hơn 5 thì dừng

```python
#Thiết lập một dict gồm các key và value
yearExperience = {
    "A" : 3,
    "B" : 4,
    "C" : 6,
    "D" : 3
}
#Thiết lập 2 biến years và b đồng thời cho các giá trị cảu chúng bằng 0
years = 0
b = 0

#Ép dict thành list
yearsList = list(yearExperience.items())

#Khi biến years nhỏ hơn 5 biến key và years sẽ nhận lần lượt "A" và 3 từ list yearList bị truy cập bởi b
while years < 5:
    key, years  = yearsList[b]

    #In giá trị của key và years ra màn hình và thực hiện phép tăng biến cho b,  b += 1 
    #Lúc này b sẽ nhận giá trị thứ 2 trong list và tiếp lục quay lại while để thực hiện vòng lặp đến khi years > 5 thì dừng
    print(key, years)
    b += 1
```

# Classes
- Class (Lớp) trong Python là một khuôn mẫu (blueprint) hoặc một định nghĩa để tạo ra các đối tượng (objects). Nó là một cấu trúc dữ liệu tổng hợp cho phép bạn đóng gói dữ liệu và các hàm liên quan thành một đơn vị.  
- Khi bạn định nghĩa một class, bạn không thực sự tạo ra dữ liệu, mà chỉ mô tả cách thức một loại đối tượng nào đó sẽ được tạo ra.
- Đồng thời Nó là khái niệm cốt lõi của Lập trình hướng đối tượng (Object-Oriented Programming - OOP).

   - Dữ liệu (Data): Được gọi là các thuộc tính (attributes), đại diện cho trạng thái của đối tượng.

   - Hàm (Functions): Được gọi là các phương thức (methods), đại diện cho hành vi của đối tượng.
- Class mang lại nhiều lợi ích quan trọng trong lập trình:

```python
class HieuList:
    def __init__(self):
        # Khởi tạo một list rỗng.
        self._items = []

    def add(self, item):
        # Thêm một phần tử vào cuối list
        self._items.append(item)

    def __getitem__(self, index):
        # Truy xuất một phần tử bằng index (chỉ mục).
       return self._items[index]

    def __setitem__(self, index, value):
        # Đặt (gán) một phần tử tại một chỉ mục cụ thể.
        self._items[index] = value

    def __repr__(self):
        # Trả về một chuỗi đại diện cho list.
        return str(self._items)

    def __len__(self):
        # Trả về độ dài của list.
        return len(self._items)
```

### Excercise: 
Đoạn mã này định nghĩa một Lớp (Class) tên là BaseSalary trong Python, được thiết kế để quản lý và tính toán các thành phần lương cơ bản, thưởng, và tổng lương cho một nhân viên.

```python
class BaseSalary:
    def __init__(self, baseSalary, bonusRate=.1, symbol="$"):
        
        # Lưu trữ các thuộc tính cơ bản
        self.baseSalary = baseSalary
        self.bonusRate = bonusRate
        self.symbol = symbol

    def __repr__(self):
        # Biểu diễn đối tượng: Ký hiệu + Lương cơ bản
        return f"{self.symbol}{self.baseSalary}"
    
    def get_total_salary(self):
        # Tính toán và trả về tổng lương (Cơ bản + Thưởng)
        total = self.baseSalary * (1 + self.bonusRate)

        # Format kết quả không có phần thập phân
        return f"{self.symbol}{total:.0f}"

    def get_bonus(self):
        # Tính toán và trả về số tiền thưởng.
        bonus = self.baseSalary * self.bonusRate

        # Format kết quả không có phần thập phân
        return f"{self.symbol}{bonus:.0f}"
```

```python
# Tham chiếu giá trị 100000 đến class BaseSalary
salary = BaseSalary(100000)
```

```python
salary
```

```python
# Khởi tạo biến salary tham số 100_000 đên hàm BaseSalary 
salary = BaseSalary(100_000)

# In lần lượt các giá trị của baseSalary, bonusRate và symbol trong class
print(salary.baseSalary)
print(salary.bonusRate)
print(salary.symbol)

# Khai triển 2 hàm tính tổng lương và tính lượng lương được bonus
print(salary.get_total_salary())
print(salary.get_bonus())
```

## Data Cleaning
Data Cleaning là giai đoạn quan trọng nhất trong quy trình Khoa học Dữ liệu, nơi dữ liệu thô được kiểm tra, sửa chữa và chuẩn bị để đảm bảo chất lượng và độ tin cậy cho quá trình phân tích.

Tác Dụng Chính Của Data Cleaning:
- Giải quyết Vấn đề Dữ liệu Thiếu
- Sử dụng Giá trị Trung vị để Điền Dữ liệu
- Tạo Bộ Dữ liệu Sẵn sàng Phân tích


```python
# Nhập thư viện Pandas để xử lý dữ liệu
import pandas as pd
# Nhập hàm để tải dataset
from datasets import load_dataset
# Nhập thư viện Matplotlib để trực quan hóa
import matplotlib.pyplot as plt

# Tải dataset 'data_jobs'
dataset = load_dataset('lukebarousse/data_jobs')
# Chuyển dữ liệu huấn luyện ('train') sang DataFrame của Pandas
df = dataset['train'].to_pandas()

# Chuyển đổi cột ngày đăng thành định dạng datetime
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Sử dụng loc để chọn các hàng từ đầu đến index 10 và tất cả các cột từ 'salary_year_avg' đến 'salary_hour_avg'
df.loc[:10,'salary_year_avg':'salary_hour_avg']
```

```python
# Tính toán giá trị trung vị (median) của tất cả các mức lương trung bình năm
medianSalaryYear = df['salary_year_avg'].median()
```

```python
# Tính toán giá trị trung vị (median) của tất cả các mức lương trung bình giờ
medianSalaryHour = df['salary_hour_avg'].median()
```

```python
# Gán df cho df_filled (tạo bản sao hoặc tham chiếu)
df_filled = df
# Điền các giá trị NaN trong cột 'salary_year_avg' bằng giá trị trung vị của cột đó
df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(medianSalaryYear)
# Điền các giá trị NaN trong cột 'salary_hour_avg' bằng giá trị trung vị của cột đó
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(medianSalaryHour)
```

```python
# Sử dụng loc trên df_filled để chọn các hàng từ đầu đến index 10 và các cột lương
df_filled.loc[:10, 'salary_year_avg':'salary_hour_avg']
```

```python
# Điền giá trị NaN trong cột lương năm bằng giá trị trung vị đã tính
df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(medianSalaryYear)
# Điền giá trị NaN trong cột lương giờ bằng giá trị trung vị đã tính
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(medianSalaryHour)
```

```python
# Sử dụng loc trên df_filled để chọn các hàng từ đầu đến index 10 và các cột lương
df_filled.loc[:10,'salary_year_avg':'salary_hour_avg']
```

```python
# Tạo DataFrame df_unique là bản sao của df_filled
df_unique = df_filled
# Loại bỏ các dòng bị trùng lặp (giữ lại dòng đầu tiên)
df_unique = df_unique.drop_duplicates()
# In ra tổng số dòng của DataFrame gốc
print('Length of original df:       ', len(df_filled))
# In ra tổng số dòng sau khi đã loại bỏ trùng lặp
print('Length of drop duplicates df:', len(df_unique))
# In ra số lượng dòng đã bị xóa
print('Rows Dropped:                ', len(df_filled)-len(df_unique))
```

## Median Salary Analysis by Country
Phân tích này là một ví dụ nâng cao về việc sử dụng pd.pivot_table() để tổng hợp dữ liệu dựa trên hai danh mục (job_country và job_title_short) và tính toán một giá trị số (salary_year_avg).
- Tiêu chuẩn hóa Mức lương Toàn cầu
- Phát hiện Sự chênh lệch
- Trực quan hóa So sánh
- Lọc và Tập trung Phân tích

```python
# Nhập thư viện Pandas để xử lý dữ liệu
import pandas as pd

# Nhập hàm để tải dataset
from datasets import load_dataset

# Nhập thư viện Matplotlib để trực quan hóa
import matplotlib.pyplot as plt

# Tải dataset 'lukebarousse/data_jobs'
dataset = load_dataset('lukebarousse/data_jobs')

# Chuyển dữ liệu huấn luyện ('train') thành Pandas DataFrame
df = dataset['train'].to_pandas()

# Chuyển đổi cột ngày đăng thành định dạng datetime
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## Merge DataFrames
Hợp nhất DataFrames (Merging DataFrames) là quá trình kết hợp hai hoặc nhiều DataFrame lại với nhau dựa trên một hoặc nhiều khóa chung (Key), tương tự như các thao tác JOIN trong SQL.

Mục tiêu chính trong ví dụ này là kết hợp một DataFrame chứa dữ liệu tổng hợp theo tháng (df_US_software_pivot) với một DataFrame khác (hoặc Series) chứa dữ liệu cần kết hợp.
Tác dụng của Merge DataFrames:
- Kết hợp Các Nguồn Dữ liệu Khác nhau
- Bổ sung Thuộc tính
- Căn chỉnh Dữ liệu Dòng thời gian
- Trực quan hóa Dữ liệu Tổng thể

```python
# Nhập thư viện Pandas để xử lý dữ liệu
import pandas as pd
# Nhập hàm để tải dataset
from datasets import load_dataset
# Nhập thư viện Matplotlib để trực quan hóa
import matplotlib.pyplot as plt

# Tải dataset 'lukebarousse/data_jobs'
dataset = load_dataset('lukebarousse/data_jobs')
# Chuyển dữ liệu huấn luyện ('train') thành Pandas DataFrame
df = dataset['train'].to_pandas()

# Chuyển đổi cột ngày đăng thành định dạng datetime
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## Applying Function
Phương thức .apply() cho phép bạn thực thi một hàm (có sẵn hoặc tự định nghĩa) lên từng giá trị (hoặc từng hàng/cột) của một Series hoặc DataFrame.
Tác dụng của Applying Function:
- Biến đổi Kiểu dữ liệu Phức tạp
- Đảm bảo Tính Đồng nhất của Dữ liệu
- Lọc Dữ liệu Thiếu có Mục đích

```python
# Nhập thư viện Pandas để xử lý dữ liệu
import pandas as pd

# Nhập hàm để tải dataset
from datasets import load_dataset

# Nhập thư viện Matplotlib để trực quan hóa
import matplotlib.pyplot as plt

# Tải dataset 'lukebarousse/data_jobs'
dataset = load_dataset('lukebarousse/data_jobs')

# Chuyển dữ liệu huấn luyện ('train') thành Pandas DataFrame
df = dataset['train'].to_pandas()

# Chuyển đổi cột ngày đăng thành định dạng datetime
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Sử dụng pd.notna để tạo điều kiện lọc, sau đó chọn cột 'salary_year_avg' với các giá trị không phải NaN
df[pd.notna(df['salary_year_avg'])]['salary_year_avg']
```

```python
# Lọc DataFrame để chỉ giữ lại các hàng có mức lương năm, và tạo bản sao độc lập
df_salary = df[pd.notna(df['salary_year_avg'])].copy()

# Định nghĩa hàm tính toán mức lương dự kiến tăng 3%
def projected_salary(salary):
    return salary * 1.03

# Áp dụng hàm dự kiến tăng lương cho cột mức lương năm và tạo cột mới
df_salary['salary_year_inflated'] = df_salary['salary_year_avg'].apply(projected_salary)

# Hiển thị cột mức lương gốc và mức lương đã được dự kiến tăng để so sánh
df_salary[['salary_year_avg', 'salary_year_inflated']]
```

```python
# Áp dụng hàm lambda để tăng mức lương năm thêm 3% và tạo cột mới
df_salary['salary_year-inflated'] = df_salary['salary_year_avg'].apply(lambda salary: salary * 1.03)
# Hiển thị cột mức lương gốc và mức lương đã được dự kiến tăng để so sánh
df_salary[['salary_year_avg', 'salary_year-inflated']]
```

- Tính toán Vector hóa (Vectorized Calculation): Áp dụng một phép tính đơn giản, đồng nhất lên toàn bộ Series bằng cách nhân một hằng số.
- Áp dụng Hàm Tùy chỉnh (.apply(..., axis=1)): Áp dụng logic phức tạp, có điều kiện, lên từng hàng của DataFrame.

Tác dụng chính của phần bài tập này là:  
   - Cho phép ước tính mức lương hoặc chi phí trong tương lai
   - Cung cấp khả năng thực thi các quy tắc kinh doanh phức tạp mà không thể thực hiện bằngcác    phép tính vector hóa đơn giản.
   - Việc định nghĩa các hàm Python tùy chỉnh giúp mã hóa các quy tắc kinh doanh một cách rõràng  và dễ đọc. Khi quy tắc thay đổi (ví dụ: mức tăng lương cho Senior thay đổi), bạn chỉcần sửa     đổi hàm projected_salary mà không cần thay đổi cấu trúc mã DataFrame lớn.

## Explode
Phương thức df.explode() là một công cụ mạnh mẽ trong Pandas dùng để biến các giá trị dạng danh sách (list) hoặc mảng (array) trong một ô thành các hàng mới riêng biệt.

Tác dụng chính của explode:
- Chuyển đổi dữ liệu từ trạng thái "một-nhiều" (một hàng có nhiều kỹ năng) thành trạng thái "một-một" (một hàng chỉ có một kỹ năng)
- Sau khi explode(), bạn có thể sử dụng .value_counts() để đếm số lần xuất hiện của từng kỹ năng riêng lẻ trên toàn bộ tập dữ liệu một cách chính xác. Nếu không có explode(), việc đếm sẽ bị sai vì nó sẽ đếm cả danh sách là một đơn vị.
- Sau khi đếm và sắp xếp, dữ liệu đã sẵn sàng để vẽ Biểu đồ Cột Ngang, giúp so sánh trực quan mức độ phổ biến của các kỹ năng trong một vai trò cụ thể

```python
# Nhập thư viện Pandas
import pandas as pd

# Dữ liệu mẫu chứa chức danh công việc và kỹ năng (đã là list)
data = {
    'job_title_short': ['Data Analyst', 'Data Scientist', 'Data Engineer'],
    'job_skills': [['excel', 'sql', 'python'], ['python', 'r'], ['aws', 'python', 'airflow']]
}

# Tạo DataFrame từ dữ liệu mẫu
df_skill_lists = pd.DataFrame(data)

# Hiển thị DataFrame
df_skill_lists
```

Tác dụng chính cảu bài tập làn này là:

- Tối ưu hóa Trực quan hóa và căn chỉnh phân bố nguồn lực

```python
# Nhập thư viện Pandas để xử lý dữ liệu
import pandas as pd
# Nhập hàm để tải dataset
from datasets import load_dataset
# Nhập thư viện Matplotlib để trực quan hóa
import matplotlib.pyplot as plt
# Nhập thư viện ast để chuyển đổi chuỗi an toàn
import ast

# Tải dataset 'lukebarousse/data_jobs'
dataset = load_dataset('lukebarousse/data_jobs')
# Chuyển dữ liệu huấn luyện ('train') thành Pandas DataFrame
df = dataset['train'].to_pandas()

# Chuyển đổi cột ngày đăng thành định dạng datetime
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
# Áp dụng hàm lambda để chuyển đổi chuỗi kỹ năng thành list, bỏ qua giá trị thiếu (NaN)
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
# Lọc DataFrame df để chỉ giữ lại các hàng có 'job_title_short' là 'Data Analyst' và sử dụng .copy() để tạo bản sao độc lập
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
```

```python
# Trích xuất số thứ tự của tháng (1-12) từ cột ngày đăng và lưu vào cột mới 'job_posted_month_no'
df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month
```

```python
# Tách mỗi kỹ năng thành một hàng riêng biệt để dễ dàng đếm
df_DA_explode = df_DA.explode('job_skills')

# Tạo bảng pivot: tháng làm index, kỹ năng làm cột, đếm số lượng tin đăng (size) và điền 0 cho ô trống
df_DA_pivot = df_DA_explode.pivot_table(
    index='job_posted_month_no',
    columns='job_skills',
    aggfunc='size',
    fill_value=0
)
```

```python
# Tính tổng số lượng tin đăng cho mỗi kỹ năng qua các tháng và thêm vào hàng mới có nhãn 'Total'
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()
# Sắp xếp các cột của bảng pivot dựa trên giá trị của hàng 'Total' (số lượng tin đăng cao nhất sang trái)
df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]
# Xóa hàng 'Total' vì đã hoàn thành nhiệm vụ sắp xếp cột
df_DA_pivot = df_DA_pivot.drop('Total')
# Hiển thị DataFrame đã sắp xếp
df_DA_pivot
```

```python
# : chọn tất cả hàng, :5 chọn từ cột 0 đến 4 không bao gồm index colume
df_DA_pivot.iloc[:, :5].plot(kind="line")

# Đặt tiêu đề cho biểu đồ
plt.title("Top 5 skills for Data Analyst pe month")
# Đặt nhãn cho trục x để trống ' '
plt.xlabel('')
# Đặt nhãn cho trục y
plt.ylabel('Count')
# Hiển thị biểu đồ
plt.show()
```

```python
# Tạo bản sao (copy) của DataFrame gốc (df)
df_skills = df.copy()

# Tách các danh sách kỹ năng trong cột 'job_skills' thành các hàng riêng biệt
df_skills = df_skills.explode('job_skills')

# Đếm số lần xuất hiện của mỗi cặp (Kỹ năng, Chức danh)
skill_count = df_skills.groupby(['job_skills', 'job_title_short']).size()

# Tạo cột mới 'skill_count' từ số đếm (size) 
df_skills_count = skill_count.reset_index(name='skill_count')

# Sắp xếp DataFrame theo cột 'skill_count' từ cao xuống thấp, 'inplace=True' áp dụng thay đổi trực tiếp lên df_skills_count
df_skills_count.sort_values(by='skill_count', ascending=False, inplace=True)

df_skills_count
```

```python
# Import thư viện cần thiết: pandas để xử lý dữ liệu, load_dataset để tải dataset từ Hugging Face, matplotlib.pyplot để vẽ biểu đồ, và ast để chuyển đổi chuỗi sang danh sách.
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import ast

# Tải dataset 'data_jobs' từ Hugging Face và chuyển đổi thành DataFrame của pandas.
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Xử lý cột 'job_skills': Chuyển đổi chuỗi ngày giờ thành kiểu dữ liệu datetime
# Dùng ast.literal_eval() an toàn để chuyển đổi, bỏ qua nếu giá trị là NaN
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
# Dùng ast.literal_eval() an toàn để chuyển đổi, bỏ qua nếu giá trị là NaN
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
# Tạo bản sao job_title_short 
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Lọc DataFrame gốc (df) để chỉ lấy các tin tuyển dụng 'Data Analyst' và tạo bản sao.
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month
# 'Explode' (phân tách) DataFrame: Mỗi tin tuyển dụng có nhiều kỹ năng sẽ được tách thành nhiều hàng, mỗi hàng chứa một kỹ năng.
df_DA_explode = df_DA.explode('job_skills')
# Tạo bảng tổng hợp (Pivot Table) để đếm số lần xuất hiện của từng kỹ năng theo từng tháng.
df_DA_pivot = df_DA_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

# Tính tổng số lần xuất hiện của MỖI KỸ NĂNG trên toàn bộ các tháng.
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()
# Sắp xếp lại các CỘT (kỹ năng) dựa trên tổng số lần xuất hiện theo thứ tự GIẢM DẦN.
df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]
# Xóa hàng tổng 'Total' sau khi đã sắp xếp.
df_DA_pivot = df_DA_pivot.drop('Total')

df_DA_pivot = df_DA_pivot.reset_index()

# Chuyển số tháng (1, 2, ...) thành tên tháng rút gọn (Jan, Feb, ...) để hiển thị đẹp hơn trên biểu đồ.
df_DA_pivot['job_posted_month'] = df_DA_pivot['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))

# Đặt tên tháng làm chỉ mục mới.
df_DA_pivot = df_DA_pivot.set_index('job_posted_month')

df_DA_pivot = df_DA_pivot.drop(columns='job_posted_month_no')

# Vẽ biểu đồ đường (line plot) cho 5 cột đầu tiên (Top 5 kỹ năng phổ biến nhất).


df_DA_pivot.iloc[:, :5].plot(
    kind='line',
    linewidth=4,
    linestyle='--',
    cmap='viridis',
    marker='o',
    markersize=5,
    figsize=(10,5)
)

plt.title('Top 5 Skills for Data Analysts per Month')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
```

- Đường Trung vị (Median): Đường kẻ bên trong hộp (Quartile 2, Q2)
- Hộp (Box): Mở rộng từ Bách phân vị thứ 25 (Q1) đến Bách phân vị thứ 75 (Q3)
- Khoảng Tứ phân vị (IQR): Chiều dài của hộp (Q3 - Q1)
- Râu (Whiskers): Các đường kéo dài từ hộp ra, thường đến mức tối đa hoặc tối thiểu không phải là giá trị ngoại lệ
- Giá trị Ngoại lệ (Outliers): Các điểm dữ liệu nằm ngoài ranh giới của râu, thường được biểu diễn bằng các chấm nhỏ

Tác dụng chính của box plots:
- Cho phép so sánh trực quan các chỉ số thống kê chính (trung vị, độ phân tán, và khoảng tứ phân vị) của nhiều nhóm dữ liệu cùng một lúc
- làm nổi bật các điểm dữ liệu bất thường (Outliers) bằng các chấm, dễ dàng nhận diện và hình dung
- Tóm tắt nhanh chóng và so sánh sự phân phối của các biến số định lượng một cách hiệu quả.

```python
# Import thư viện cần thiết: pandas để xử lý dữ liệu, load_dataset để tải dataset từ Hugging Face, matplotlib.pyplot để vẽ biểu đồ, và ast để chuyển đổi chuỗi sang danh sách.
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import ast
from matplotlib.ticker import FuncFormatter

# Tải dataset 'data_jobs' từ Hugging Face và chuyển đổi thành DataFrame của pandas.
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Xử lý cột 'job_skills': Chuyển đổi chuỗi ngày giờ thành kiểu dữ liệu datetime
# Dùng ast.literal_eval() an toàn để chuyển đổi, bỏ qua nếu giá trị là NaN
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
# Dùng ast.literal_eval() an toàn để chuyển đổi, bỏ qua nếu giá trị là NaN
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

- Giúp hiểu rõ kỹ năng nào là cốt lõi (nhu cầu cao) và kỹ năng nào là chuyên môn (lương cao), từ đó định giá và tìm kiếm ứng viên tốt hơn

```python
# Tách cột 'job_skills' thành các hàng riêng biệt để mỗi hàng chỉ chứa một kỹ năng
df_skills = df_US.explode('job_skills')

# Hiển thị các cột 'job_title' và 'job_skills' để kiểm tra kết quả tách
df_skills[['job_title', 'job_skills']]
```

```python
# Nhóm DataFrame theo kỹ năng và chức danh công việc để đếm số lần xuất hiện của mỗi cặp
df_skills_count = df_skills.groupby(['job_skills', 'job_title_short']).size()
# Chuyển Series MultiIndex thành DataFrame và đặt tên cột đếm là 'skill_count'
df_skills_count = df_skills_count.reset_index(name='skill_count')
# Sắp xếp DataFrame theo số lượng kỹ năng giảm dần (inplace=True)
df_skills_count.sort_values(by='skill_count', ascending=False, inplace=True)
# Hiển thị DataFrame
df_skills_count
```

```python
# Trích xuất danh sách tất cả các chức danh công việc duy nhất từ DataFrame
job_titles = df_skills_count['job_title_short'].unique().tolist()
# Chọn 3 chức danh đầu tiên (từ danh sách gốc) và sắp xếp chúng theo thứ tự bảng chữ cái
job_titles = sorted(job_titles[:3])
# Hiển thị danh sách chức danh
job_titles
```

```python
# Tạo figure với số lượng subplot bằng số lượng job_titles
fig, ax = plt.subplots(len(job_titles), 1)

# Áp dụng theme Seaborn
sns.set_theme(style='ticks')

# Bắt đầu vòng lặp qua từng chức danh công việc để tạo subplot
for i, job_title in enumerate(job_titles):
    # Lọc, lấy 5 kỹ năng hàng đầu, và đảo ngược thứ tự cho việc vẽ biểu đồ cột ngang
    df_plot = df_skills_count[df_skills_count['job_title_short'] == job_title].head(5)[::-1]
    # Vẽ biểu đồ cột ngang (barplot) lên trục hiện tại, sử dụng màu sắc theo count
    sns.barplot(data=df_plot, x='skill_count', y='job_skills', ax=ax[i], hue='skill_count', palette='dark:b_r')
    
    # Các tùy chỉnh biểu đồ cơ bản
    ax[i].set_title(job_title)
    ax[i].invert_yaxis()
    ax[i].set_ylabel('')
    ax[i].set_xlabel('')
    ax[i].get_legend().remove()
    # Đồng bộ hóa giới hạn trục X
    ax[i].set_xlim(0, 45000)

# Đặt tiêu đề chung cho figure và điều chỉnh layout
fig.suptitle('Counts of Skills Requested in US Job Postings', fontsize=15)
fig.tight_layout(h_pad=0.5)
plt.show()
```

```python
# Đếm tần suất xuất hiện của mỗi chức danh công việc trong df_US và đặt tên cột đếm là 'jobs_total'
df_job_title_count = df_US['job_title_short'].value_counts().reset_index(name='jobs_total')
# Hiển thị DataFrame
df_job_title_count
```

```python
# Gộp DataFrame df_skills_count (số lượng kỹ năng) và df_job_title_count (tổng số tin đăng) dựa trên tên chức danh công việc
df_skills_perc = pd.merge(df_skills_count, df_job_title_count, on='job_title_short', how='left')
# Tính tỷ lệ phần trăm của mỗi kỹ năng trên tổng số tin đăng của chức danh đó
df_skills_perc['skill_percent'] = (df_skills_perc['skill_count'] / df_skills_perc['jobs_total']) * 100
# Hiển thị DataFrame
df_skills_perc
```

```python
# Tạo figure với số lượng subplot bằng số lượng job_titles
fig, ax = plt.subplots(len(job_titles), 1)

# Áp dụng theme Seaborn
sns.set_theme(style='ticks')

# Bắt đầu vòng lặp qua từng chức danh công việc để tạo subplot
for i, job_title in enumerate(job_titles):
    # Lọc DataFrame để lấy 5 kỹ năng hàng đầu (theo count)
    df_plot = df_skills_perc[df_skills_perc['job_title_short'] == job_title].head(5)
    # Vẽ biểu đồ cột ngang (barplot) với x là tỷ lệ phần trăm (skill_percent)
    sns.barplot(data=df_plot, x='skill_percent', y='job_skills', ax=ax[i], hue='skill_count', palette='dark:b_r')
    
    # Các tùy chỉnh cơ bản
    ax[i].set_title(job_title)
    ax[i].set_ylabel('')
    ax[i].set_xlabel('')
    ax[i].get_legend().remove()
    ax[i].set_xlim(0, 78) 
    
    # Ẩn nhãn tick trục X nếu không phải biểu đồ cuối cùng
    if i != len(job_titles) - 1:
        ax[i].set_xticks([])

    # Vòng lặp để gắn nhãn tỷ lệ phần trăm vào cuối mỗi cột
    for n, v in enumerate(df_plot['skill_percent']):
        ax[i].text(v + 1, n, f'{v:.0f}%', va='center')

# Đặt tiêu đề chung cho figure và điều chỉnh layout
fig.suptitle('Likelihood of Skills Requested in US Job Postings', fontsize=15)
fig.tight_layout(h_pad=.8)
plt.show()
```

```python
# Import thư viện cần thiết: pandas để xử lý dữ liệu, load_dataset để tải dataset từ Hugging Face, matplotlib.pyplot để vẽ biểu đồ, và ast để chuyển đổi chuỗi sang danh sách.
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import seaborn as sns
import ast
from matplotlib.ticker import FuncFormatter

# Tải dataset 'data_jobs' từ Hugging Face và chuyển đổi thành DataFrame của pandas.
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Xử lý cột 'job_skills': Chuyển đổi chuỗi ngày giờ thành kiểu dữ liệu datetime
# Dùng ast.literal_eval() an toàn để chuyển đổi, bỏ qua nếu giá trị là NaN
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
# Dùng ast.literal_eval() an toàn để chuyển đổi, bỏ qua nếu giá trị là NaN
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
# Lọc DataFrame để chỉ giữ lại công việc 'Data Analyst' ở Mỹ và tạo bản sao độc lập
df_DA_US = df[(df['job_title'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
# Tạo cột số thứ tự tháng (1-12) từ ngày đăng tin
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month
# Tách từng kỹ năng trong list thành hàng riêng biệt và lưu vào DataFrame mới
df_DA_US_explode = df_DA_US.explode('job_skills')
```

```python
# Tạo một pivot table với index là job_posted_month_no và cột là job_skills
df_DA_US_pivot = df_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills',  aggfunc='size', fill_value=0)

df_DA_US_pivot
```

```python
# Tính tổng số lượng tin đăng cho mỗi kỹ năng qua các tháng và thêm vào hàng mới có nhãn 'Total'
df_DA_US_pivot.loc['Total'] = df_DA_US_pivot.sum()

# Sắp xếp các cột của bảng pivot dựa trên giá trị của hàng 'Total' (số lượng tin đăng cao nhất sang trái)
df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index]

# Xóa hàng 'Total' vì đã hoàn thành nhiệm vụ sắp xếp cột
df_DA_US_pivot = df_DA_US_pivot.drop('Total')

# Hiển thị DataFrame đã sắp xếp
df_DA_US_pivot
```

```python
# Chọn tất cả các hàng và 5 cột đầu tiên (5 kỹ năng phổ biến nhất) rồi vẽ biểu đồ đường
df_DA_US_pivot.iloc[:, :5].plot(kind='line')
# Đặt tiêu đề, nhãn trục Y, X và hiển thị biểu đồ
plt.title('Trending Top Skills for Data Analysts in the US')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
```

```python
# Nhóm DataFrame theo số thứ tự tháng và tính số lượng tin đăng (size) trong mỗi tháng
DA_totals = df_DA_US.groupby('job_posted_month_no').size()
# Hiển thị kết quả tổng số tin đăng theo tháng
DA_totals
```

```python
# Chia số lượng tin đăng mỗi kỹ năng cho tổng số tin đăng của tháng (DA_totals/100) để tính tỷ lệ phần trăm
df_DA_US_percent = df_DA_US_pivot.iloc[:12].div(DA_totals/100, axis=0)
# Đặt lại index (cột số thứ tự tháng) thành cột dữ liệu
df_DA_US_percent = df_DA_US_percent.reset_index()
# Chuyển đổi số thứ tự tháng (1-12) thành tên viết tắt của tháng (Jan, Feb,...)
df_DA_US_percent['job_posted_month'] = df_DA_US_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
# Đặt lại tên tháng làm index mới
df_DA_US_percent = df_DA_US_percent.set_index('job_posted_month')
# Xóa cột số thứ tự tháng tạm thời
df_DA_US_percent = df_DA_US_percent.drop(columns='job_posted_month_no')
# Hiển thị DataFrame đã tính toán
df_DA_US_percent
```

```python
# Import thư viện cần thiết: pandas để xử lý dữ liệu, load_dataset để tải dataset từ Hugging Face, matplotlib.pyplot để vẽ biểu đồ, và ast để chuyển đổi chuỗi sang danh sách.
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import seaborn as sns
import ast
from matplotlib.ticker import FuncFormatter

# Tải dataset 'data_jobs' từ Hugging Face và chuyển đổi thành DataFrame của pandas.
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Xử lý cột 'job_skills': Chuyển đổi chuỗi ngày giờ thành kiểu dữ liệu datetime
# Dùng ast.literal_eval() an toàn để chuyển đổi, bỏ qua nếu giá trị là NaN
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
# Dùng ast.literal_eval() an toàn để chuyển đổi, bỏ qua nếu giá trị là NaN
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
# Lọc DataFrame để chỉ giữ lại các công việc 'Data Analyst' ở 'United States' và tạo bản sao độc lập
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()

# Loại bỏ các hàng có giá trị NaN trong cột mức lương trung bình năm (để đảm bảo tính toán chính xác)
df_DA_US = df_DA_US.dropna(subset=['salary_year_avg'])

# Tách mỗi kỹ năng trong list thành một hàng riêng biệt, nhân bản các dữ liệu khác
df_DA_US_exploded = df_DA_US.explode('job_skills')

# Hiển thị 5 hàng đầu tiên của cột mức lương và kỹ năng để kiểm tra
df_DA_US_exploded[['salary_year_avg', 'job_skills']].head(5)
```

```python
# Nhóm dữ liệu theo kỹ năng, tính số lượng tin đăng và mức lương trung vị, sắp xếp theo số lượng giảm dần
df_DA_skills = df_DA_US_exploded.groupby('job_skills')['salary_year_avg'].agg(['count', 'median']).sort_values(by='count', ascending=False)

# Đổi tên cột đếm và cột trung vị để dễ đọc
df_DA_skills = df_DA_skills.rename(columns={'count': 'skill_count', 'median': 'median_salary'})

# Tính tổng số lượng tin tuyển dụng Data Analyst 
DA_job_count = len(df_DA_US)

# Tính tỷ lệ phần trăm của mỗi kỹ năng trên tổng số tin tuyển dụng
df_DA_skills['skill_percent'] = df_DA_skills['skill_count'] / DA_job_count * 100

# Đảm bảo chỉ giữ lại các hàng có số lượng kỹ năng lớn hơn 0
df_DA_skills = df_DA_skills[df_DA_skills['skill_count'] > 0]

# Hiển thị DataFrame
df_DA_skills
```

```python
# Đặt ngưỡng tỷ lệ phần trăm là 5
skill_limit = 5

# Lọc DataFrame df_DA_skills để giữ lại các hàng có 'skill_percent' lớn hơn ngưỡng
df_DA_skills_high_demand = df_DA_skills[df_DA_skills['skill_percent'] > skill_limit]
```

```python
# Nhập thư viện để điều chỉnh vị trí nhãn văn bản
from adjustText import adjust_text

# Vẽ biểu đồ phân tán (scatter plot)
plt.scatter(df_DA_skills_high_demand['skill_percent'], df_DA_skills_high_demand['median_salary'])
# Đặt nhãn trục X và Y, và tiêu đề
plt.xlabel('Percent of Data Analyst Jobs')
plt.ylabel('Median Salary ($USD)') 
plt.title('Most Optimal Skills for Data Analysts in the US')

# Lấy trục hiện tại và định dạng nhãn trục Y thành đơn vị nghìn đô ($XK)
ax = plt.gca()
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, pos: f'${int(y/1000)}K')) 

# Vòng lặp để tạo nhãn văn bản cho từng điểm dữ liệu
texts = []
for i, txt in enumerate(df_DA_skills_high_demand.index):
    texts.append(plt.text(df_DA_skills_high_demand['skill_percent'].iloc[i], df_DA_skills_high_demand['median_salary'].iloc[i], " " + txt))

# Điều chỉnh vị trí của các nhãn văn bản để tránh chồng chéo và thêm mũi tên
adjust_text(texts, arrowprops=dict(arrowstyle='->', color='gray'))

# Hiển thị biểu đồ
plt.show()
```

```python
# Lấy bản sao của cột 'job_type_skills'
df_technology = df['job_type_skills'].copy()

# Loại bỏ các dòng bị trùng lặp
df_technology = df_technology.drop_duplicates()

# Loại bỏ các giá trị bị thiếu (NaN)
df_technology = df_technology.dropna()

# Khởi tạo dictionary chính để gộp dữ liệu
technology_dict = {}
# Bắt đầu vòng lặp để gộp tất cả các dictionary lại với nhau
for row in df_technology:
    # Chuyển chuỗi thành dictionary Python
    row_dict = ast.literal_eval(row)  
    for key, value in row_dict.items():
        # Nếu key đã tồn tại, nối thêm danh sách kỹ năng
        if key in technology_dict:  
            technology_dict[key] += value
        # Nếu key chưa tồn tại, thêm key và value
        else:  
            technology_dict[key] = value

# Vòng lặp để loại bỏ các kỹ năng trùng lặp trong từng danh sách (bằng cách chuyển sang set rồi lại về list)
for key, value in technology_dict.items():
    technology_dict[key] = list(set(value))
```

```python
# Chuyển dictionary 'technology_dict' thành DataFrame, đặt tên cột là 'technology' và 'skills'
df_technology = pd.DataFrame(list(technology_dict.items()), columns=['technology', 'skills'])

# Tách mỗi kỹ năng trong list thành một hàng riêng biệt, nhân bản các giá trị 'technology'
df_technology = df_technology.explode('skills')

# Hiển thị DataFrame sau khi xử lý
df_technology
```

```python
# Gộp df_DA_skills và df_technology, sử dụng cột 'job_skills' và 'skills' làm khóa gộp
df_DA_skills_tech = df_DA_skills.merge(df_technology, left_on='job_skills', right_on='skills')

# Hiển thị DataFrame đã gộp
df_DA_skills_tech
```

```python
# Lọc df_DA_skills_tech để giữ lại các kỹ năng có 'skill_percent' lớn hơn giá trị của biến skill_limit (ngưỡng 5%)
df_DA_skills_tech_high_demand = df_DA_skills_tech[df_DA_skills_tech['skill_percent'] > skill_limit]
```

```python
# Vẽ biểu đồ phân tán: x là tỷ lệ phổ biến, y là mức lương trung vị, màu sắc (hue) phân biệt theo loại công nghệ
sns.scatterplot(
    data=df_DA_skills_tech_high_demand,
    x='skill_percent',
    y='median_salary',
    hue='technology'
)

# Áp dụng theme Seaborn và xóa đường viền trên và phải của biểu đồ
sns.despine()
sns.set_theme(style='ticks')

# Lặp qua dữ liệu để tạo nhãn văn bản cho từng điểm kỹ năng
texts = []
for i, txt in enumerate(df_DA_skills_tech_high_demand.index):
    texts.append(plt.text(df_DA_skills_tech_high_demand['skill_percent'].iloc[i], df_DA_skills_tech_high_demand['median_salary'].iloc[i], txt))

# Điều chỉnh vị trí của các nhãn văn bản để tránh chồng chéo và thêm mũi tên
adjust_text(texts, arrowprops=dict(arrowstyle='->', color='gray'))

# Đặt nhãn trục và tiêu đề biểu đồ
plt.xlabel('Percent of Data Analyst Jobs')
plt.ylabel('Median Yearly Salary')
plt.title('Most Optimal Skills for Data Analysts in the US')
# Đặt tiêu đề cho chú giải
plt.legend(title='Technology')

# Định dạng trục Y thành đơn vị nghìn đô và trục X thành tỷ lệ phần trăm
from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, pos: f'${int(y/1000)}K'))
ax.xaxis.set_major_formatter(PercentFormatter(decimals=0))

# Tự động điều chỉnh layout và hiển thị biểu đồ
plt.tight_layout()
plt.show()
```
